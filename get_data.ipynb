```python
# Puts a certain number of data from top companies into the data folder
# This includes a company facts file, and daily OHLCV data
```

```python
import datetime
import os
import scraper
import json
```

```python
NUM_COMPANIES = 500
```

```python
if not os.path.exists("./data"):
    os.mkdir("./data")
cur_datetime_str = datetime.datetime.now().__str__().replace(" ", "-").replace(":", "-")
rmi = cur_datetime_str.index(".", 1)
cur_datetime_str = cur_datetime_str[:rmi]
```

```python
data_path = f"./data/{datetime.datetime.now().__str__().replace(" ", "-").replace(":", "-")}"
os.mkdir(data_path)
```

```python
companies = scraper.get_companies()
```

```python
for i in range(NUM_COMPANIES):
    comp = companies[i]
    company_data_path = f"{data_path}/{comp.cik}"
    if (not os.path.exists(company_data_path)):
        os.mkdir(company_data_path)

    # Get company facts data
    if not os.path.exists(f"{company_data_path}/companyfacts.json"):
        with open(f"{company_data_path}/companyfacts.json", 'w') as file:
            try:
                facts = comp.get_facts()
                file.write(json.dumps(facts, indent=4))
            except RuntimeError: # some cik's don't have company facts (e.g. s&p 500 trust)
                pass
    
    # TODO: Get OHLCV data

    break
```
